```python
#functions inside format

def pct_change(x1,x2): 
    #x2 is newest 
    return np.round(((x2-x1)/np.abs(x1)) *100,2)

def x_1change(a): 
    if len(a) >= 2: 
        return pct_change(a[1], a[0])
    else: 
        return np.nan
    
def x_2change(a): 
    if len(a) >= 3: 
        return pct_change(a[2], a[0])
    else: 
        return np.nan
        
```

```python
def format_df_date(df_dict): 
    
    df_dic = defaultdict(list)
    not_done = []
    for key in df_dict.keys():
        if len(df_dict[key]) == 0: 
            not_done.append(key)
        else:
            df_dic['symbol'].append(key)
            for column in df_dict[key].index: 
                
                if column == 'symbol': 
                    pass
                else:
                    #chnage into floats 
                    x = [float(x) for x in df_dict[key].loc[column].values if x != '']
                    if len(x) == 0: 
                        x = [np.nan for x in range(4)]

                    #calc needed values 
                    x_latest = x[0]
                    x_change1 = x_1change(x)
                    x_change2 = x_2change(x)
                    x_avg = np.mean([x for x in x if x != 0])
                    #make new columns 
                    df_dic[column].append(x_latest)
                    df_dic[column + '_change1'].append(x_change1)
                    df_dic[column + '_change2'].append(x_change2)
                    df_dic[column + '_avg'].append(x_avg)
    
    return pd.DataFrame(df_dic).set_index('symbol'), not_done

            
    
```

```python
#renaming csh columns 
def rename_csh(df, lst): 
    
    
    df1 = df.rename(columns = {element: re.sub(r'^Cash per Share', 'c/sh',element)
                             for element in lst})
    
    df1 = df1.rename(columns = {element: re.sub(r'^Free Cash Flow per Share', 'fcf/sh',element)
                             for element in lst})
    
    df1 = df1.rename(columns = {element: re.sub(r'^operatingCashFlowPerShare', 'ocf/sh',element)
                             for element in lst})
    
    
    df1 = df1.rename(columns = {element: re.sub(r'^operatingCashFlowSalesRatio', 'ocf/sales',element)
                             for element in lst})
    
    df1 = df1.rename(columns = {element: re.sub(r'^Dividend Yield', 'yld',element)
                             for element in lst})
    
    df1 =  df1.rename(columns = {element: re.sub(r'^.*Total Debt', 'debt',element)
                             for element in lst})
    df1 =  df1.rename(columns = {element: re.sub(r'^.*rice', 'price',element)
                             for element in lst})
    df1 = df1.rename(columns = lambda x: x.replace(' ', '_'))
    
    df1 = df1.rename(str.lower, axis ='columns')
    
    return df1 
    
```

```python
#filtering part 2: rev/csh 
#take each chunk and re-sort based on key metrics 
def main_rank(df): 
    
    dic = defaultdict(list)
  
    prce = 'price <= 10'
    rtrn = '(rtrn_avg50 > 5E03 or rtrn_avg > 10E03)'
    eps = '(eps >= -3)'
    pct = '(pct_change50 <= -30)'
    yld = 'yld > 0'
    a = '({eps} and {rtrn}) or ({rtrn} and ({prce} or {yld}) and {pct})'
    a = a.format(eps= eps,rtrn = rtrn,prce = prce, pct = pct,yld = yld)
    
    b = '{rtrn} and ({prce} or {eps} or {yld})'.format(eps= eps,rtrn = rtrn,prce = prce, 
                                                       pct = pct,yld = yld)
    s = '{eps} and {rtrn} and {prce} and ({yld} or {pct})'
    s = s.format(eps= eps,rtrn = rtrn,prce = prce, pct = pct,yld = yld)
    
    
    s_list = df.query(s).index.tolist()
    a_list = df.query(a).index.tolist()
    b_list = df.query(b).index.tolist()
    dic['b'] = [x for x in df.index.tolist() if x not in b_list]
    dic['a'] = [x for x in a_list if x not in s_list]
    dic['s'] = s_list 
    
    
    dic2 = defaultdict(str)
    for comp in df.index.tolist(): 
        if comp in dic['s']: 
            dic2[comp]  = 's'
        elif comp in dic['a']: 
            dic2[comp] = 'a'
        else: 
            dic2[comp] = 'b'
        
    return dic2
        
    
```

```python
def income_filter(df): 
    dict_income = defaultdict(int)
    for symbol in df.index.tolist():
        
        dict_income[symbol]
        dict_income[symbol] = 0
        ser = df.loc[symbol,'revenue':]
        if (ser['revenue_change'] > 0) or (ser['revenue'] > ser['revenue_avg']):
            dict_income[symbol] += 1
            
        if ser['revenue_change'] > 3: 
            
            dict_income[symbol] += 2

        
        if ser['grossProfitRatio'] > 0.2:
            dict_income[symbol] += 1
 
        if ser['costAndExpenses'] < ser['costAndExpenses_avg']: 
           dict_income[symbol] += 1

        if ser['costAndExpenses_change'] < ser['revenue_change']: 
            dict_income[symbol] += 1

        if ser['operatingIncomeRatio'] >= 0.1: 
            
            dict_income[symbol] += 1
        if ser['netIncome'] > ser['netIncome_avg'] or ser['netIncome_change'] > 5 or ser['netIncomeRatio'] > 0:
            dict_income[symbol] += 1
        if ser['netIncomeRatio'] >= 0.1:
            dict_income[symbol] += 2 
        if ser['researchAndDevelopmentExpenses_change'] > 0 or ser['researchAndDevelopmentExpenses'] > ser['researchAndDevelopmentExpenses_avg']: 
            dict_income[symbol] += 1
        if ser['eps_change'] > 0 or ser['eps'] > ser['eps_avg']: 
            dict_income[symbol] += 1
        if ser['tbv/sh'] > ser['tbv/sh_avg']: 
            dict_income[symbol] += 1
        if ser['tangible_asset_value'] > 0 or ser['net_current_asset_value'] >0: 
            dict_income[symbol] += 1
        if ser['income_quality'] > 1: 
            dict_income[symbol] += 1 
        if ser['working_capital'] > 0: 
            dict_income[symbol] += 1 
        if ser['days_of_inventory_on_hand'] >= 7: 
            dict_income[symbol] += 1 
        if ser['daysofsalesoutstanding'] <= 7 or ser['daysofinventoryoutstanding'] <=7: 
            dict_income[symbol] += 1 
        if ser['ern_yld'] > 0: 
            dict_income[symbol] += 1
        
        if ser['ebitda'] >0.2: 
            dict_income[symbol] += 1
    dict_income_final = defaultdict(str)
            
    for symbol, count in dict_income.items(): 
        if count >= 12: 
            dict_income_final[symbol] = 's'
        elif count >= 10: 
            dict_income_final[symbol] = 'a'
        elif count >=6: 
            dict_income_final[symbol] = 'b'
        else: 
            dict_income_final[symbol] = 'c'
    
          
        
    df['rnk_income'] = dict_income_final.values()        
    return df
        
    
```

```python
def ratio_filter(df): 
    
    dict_ratio = defaultdict(float)
    for symbol in df.index.tolist():
        
        dict_ratio[symbol]
        dict_ratio[symbol] = 0
        ser = df.loc[symbol,:]
        
        if ser['p/e'] <= ser['earn_growth1']/2: 
            dict_ratio[symbol] +=2
        if ser['p/e'] < ser['p/e_avg']: 
            dict_ratio[symbol] +=1
        if ser['p/eg'] <= 1: 
            dict_ratio[symbol] +=2
        if ser['p/e'] <1: 
            dict_ratio[symbol] +=1
        if ser['p/s'] <= 2:
            if ser['p/s'] <=1.5:
                dict_ratio[symbol] +=2
            else:
                dict_ratio[symbol] +=1
        if ser['p/s'] < ser['p/s_avg']: 
            dict_ratio[symbol] +=1
        if ser['p/b'] <= 1: 
            if ser['p/b'] <= 0.66: 
                dict_ratio[symbol] +=2
            else: 
                dict_ratio[symbol] +=1
        if ser['current_ratio'] >= 2: 
            dict_ratio[symbol] +=1
        if ser['quick_ratio'] >=0.5: 
            dict_ratio[symbol] +=2 
        if ser['p/cf'] <=15: 
            if ser['p/cf'] <=10: 
                dict_ratio[symbol] +=2
            else: 
                dict_ratio[symbol] +=1
        if ser['p/cf'] < ser['p/cf_avg']: 
            dict_ratio[symbol] +=1
        if ser['priceFairValue'] >= ser['price']: 
            dict_ratio[symbol] +=2
        if ser['ROA'] >= 0.1:
            dict_ratio[symbol] +=1
        if ser['ROE'] >= 0.2: 
            dict_ratio[symbol] +=1
        if ser['debt/eq'] <= 1 and ser['debt/eq'] >0: 
            dict_ratio[symbol] +=1
        if ser['asset_turnover'] >=1: 
            dict_ratio[symbol] +=1
            
            
        
        
            
      
    
    
    
    dict_ratio_final = defaultdict(str)
            
    for symbol, count in dict_ratio.items(): 
        if count >= 13: 
            dict_ratio_final[symbol] = 's'
        elif count >= 10: 
            dict_ratio_final[symbol] = 'a'
        elif count >=8: 
            dict_ratio_final[symbol] = 'b'
        else: 
            dict_ratio_final[symbol] = 'c'
            
    df['rnk_ratio'] = dict_ratio_final.values()        
    return df
    
    
    
```
